```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Uses RTDS scheduler, full CPU, no extra time, and no extra configuration on Flink
scheduler = 'rtds'
B = '10000'
P = '10000'
filename = '../latencies/flink_kafka_default_latency.txt'
```

```python
latencies = np.loadtxt(filename)
```

```python
normalized = (latencies)/2.1
print(np.mean(normalized))
```

```python
ms = normalized/1e6
print(np.mean(ms))
```

```python
plt.boxplot(ms)
plt.ylim(0, 150)
```

```python
plt.scatter(range(latencies.shape[0]), ms)
plt.ylim(0, 250)
plt.xlabel('Sample Number')
plt.ylabel('Latency (ms)')
```

```python
print(np.max(ms))
```

```python
import statsmodels.api as sm

#subsample = np.random.choice(ms, 2, replace=False)
#subsample = ms
subsample = sorted(ms) # They have to be sorted lol
ecdf_f = sm.distributions.ECDF(subsample)
y_f = ecdf_f(subsample)
plt.plot(subsample, y_f, linestyle='--', drawstyle='steps', clip_on=False, linewidth=2.0)
plt.xlim(0, np.max(subsample))
plt.ylim(0, 1)
plt.xlabel('Latency (ms)')
plt.ylabel('Cumulative Probability')
plt.title('ECDF')
```

```python
# How do we set budget and period?
c = []
t = []
for i in range(ms.shape[0]):
    c.append(ms[i])
    t.append(100)

total = 0
for i in range(len(c)):
    total += (c[i]/t[i])
    
U = total
P = 0.1
B = U*P
print(B)
```
